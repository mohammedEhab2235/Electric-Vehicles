# Data Cleaning

```python
df.dropna(how='all')
df
```

```python
df.drop('VIN (1-10)',axis=1,inplace=True)
df.drop('Legislative District',axis=1,inplace=True)
df.drop('2020 Census Tract',axis=1,inplace=True)
df.drop('Vehicle Location',axis=1,inplace=True)
df
```

```python
df['Electric Vehicle Type']=df['Electric Vehicle Type'].str.replace('Battery Electric Vehicle (BEV)','BEV')
df['Electric Vehicle Type']=df['Electric Vehicle Type'].str.replace('Plug-in Hybrid Electric Vehicle (PHEV)','PHEV')
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility']=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Eligibility unknown as battery range has not been researched','Unknown')
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility']=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Clean Alternative Fuel Vehicle Eligible','Eligible')
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility']=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].str.replace('Not eligible due to low battery range','Not Eligible')
df
```

```python
df.groupby('Make')[['Electric Range','Base MSRP']].agg(['count','mean']).round(2)
```

```python
df.pivot_table(index='Model Year',columns='Make',aggfunc='count').fillna(0)
```

```python
df['Make'].value_counts().plot(kind='hist')
```

```python
plt.figure(figsize=(16,6))
plt.scatter(df['Make'],df['Base MSRP'],color='b')
plt.title('Car Make by MSRP', fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
Cars_No_Modelyear= df['Model Year'].value_counts()
plt.bar(Cars_No_Modelyear.index,Cars_No_Modelyear.values,color='#2d3250')
plt.xlabel('Model Year')
plt.ylabel('Cars Number')
plt.title('Cars Number by Model Year',fontsize=14)
plt.grid(axis='y',color='g',alpha=0.35,linestyle='--')
plt.show()
```

```python
CAFV=df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts()
mylabels=CAFV.values/CAFV.values.sum()*100
plt.figure(figsize=(6,6))
plt.pie(CAFV,startangle=90,labels=mylabels.round(1),colors=['#2d3250','#f6b17a','#CD5C5C'])
plt.title('CAFV Eligibility',fontsize=14)
plt.legend(labels=CAFV.index,loc='upper left')
plt.show()
```

```python
Electric_Vehicle_Type=df['Electric Vehicle Type'].value_counts()
mylabels=Electric_Vehicle_Type.values/Electric_Vehicle_Type.values.sum()*100
plt.figure(figsize=(6,6))
plt.pie(Electric_Vehicle_Type,startangle=90,labels=mylabels.round(1),colors=['#2d3250','#f6b17a'])
plt.title('Electric Vehicle Type',fontsize=14)
plt.legend(labels=Electric_Vehicle_Type.index,loc='upper left')
plt.show()
```
